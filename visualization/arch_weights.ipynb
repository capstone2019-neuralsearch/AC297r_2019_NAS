```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
norm = np.load('/Users/dylanrandle/Downloads/experiments/search-GZ-05-ADAM-20191102-100808/normal_010.npy')
redu = np.load('/Users/dylanrandle/Downloads/experiments/search-GZ-05-ADAM-20191102-100808/reduce_010.npy')
```

```python
norm.flatten()
```

```python
ax = sns.distplot(norm.flatten(), label='normal', norm_hist=False)
ax.set_title('Galaxy Zoo Architecture Weights Distribution')

ax2 = sns.distplot(redu.flatten(), label='reduce', norm_hist=False)
# ax2.set_title('Galaxy Zoo Reduce Architecture Weights Distribution');
ax2.legend()
ax2.set_xlabel('Weight value')
ax2.set_ylabel('Count')
ax2.set_xlim([0,1])
plt.savefig('/Users/dylanrandle/Downloads/galaxy_zoo_weight_dist.png');
```

```python
a = np.concatenate((norm, redu)).flatten()
np.mean(a), np.std(a)
```
